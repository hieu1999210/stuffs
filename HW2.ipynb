#Part a) load and split data

```
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn as skl
import numpy as np 
from sklearn import datasets, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.utils.multiclass import unique_labels

RANDOM_STATE = 8
C = 100.0
GAMMA = 0.1
```

```
# load the data

#X, y = load_iris(return_X_y=True)
iris = datasets.load_iris()
data = iris['data']
labels = iris['target']
class_names = iris['target_names']
# print(iris)
```

```
# split the data into training and test sets

data_train, data_test, label_train, label_test = train_test_split(data, labels, test_size=0.50, random_state=0)
print(label_test[0])
```

```
# plot confusion matrix

def plot_confusion_matrix(y_true, y_pred, classes=class_names,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    accuracy__score = accuracy_score(y_true, y_pred)
    print("Accuracy score: %.2f" % accuracy__score)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

#Part b) Training data using softmax regression

```
clf_1 = LogisticRegression(penalty = 'l2',
                         dual = False,
                         random_state=RANDOM_STATE,
                         solver='lbfgs',
                         tol = 1e-4,
                         C = C,
                         fit_intercept = True,
                         max_iter = 1000,
                         verbose = 1,
                         multi_class='multinomial')
clf_1.fit(data, labels)
```

```
label_predict_1 = clf_1.predict(data_test)

# Plot non-normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_1, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.show()

# Plot normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_1, classes=class_names,
                      normalize=True, title='Normalized confusion matrix')
plt.show()
```

```
train_pred = clf_1.predict(data_train)
plot_confusion_matrix(label_train, train_pred, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.show()
```

#Part c) classification by SVM and linear kernel

```
clf_2 = svm.SVC(gamma=GAMMA,
                C = C,
                kernel = 'linear',
                random_state = RANDOM_STATE,
                verbose = True)
clf_2.fit(data_train, label_train)
```

```
label_predict_2 = clf_2.predict(data_test)

# Plot non-normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_2, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.show()

# Plot normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_2, classes=class_names,
                      normalize=True, title='Normalized confusion matrix')
plt.show()
```

#Part d) classification by SVM and poly kernel

```
clf_3 = svm.SVC(gamma=GAMMA,
                kernel = 'poly',
                C = C,
                random_state = RANDOM_STATE,
                verbose = True)
clf_3.fit(data_train, label_train)
```

```
label_predict_3 = clf_3.predict(data_test)

plot_confusion_matrix(label_test, label_predict_3, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.show()

# Plot normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_3, classes=class_names,
                      normalize=True, title='Normalized confusion matrix')
plt.show()
```

#Part e) classification by SVM and rbf kernel

```
clf_4 = svm.SVC(gamma=GAMMA,
                C = C,
                kernel = 'rbf',
                random_state = RANDOM_STATE,
                verbose = True)
clf_4.fit(data_train, label_train)
```

```
label_predict_4 = clf_4.predict(data_test)

plot_confusion_matrix(label_test, label_predict_4, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.show()

# Plot normalized confusion matrix
plot_confusion_matrix(label_test, label_predict_4, classes=class_names,
                      normalize=True, title='Normalized confusion matrix')
plt.show()
```

```
# Summary of results
print("Score of softmax:    %.2f" % clf_1.score(data_test, label_test))
print("Score of svm linear: %.2f" % clf_2.score(data_test, label_test))
print("Score of svm poly:   %.2f" % clf_3.score(data_test, label_test))
print("Score of svm rbf:    %.2f" % clf_4.score(data_test, label_test))
```
